```python
# we can import our libraries
import numpy as np # NB may need to pip install numpy
```

```python
# numpy offers some additional data structures. The arange, the array etc.
# arange (array-range) is like range but makes an array
n = np.arange(2, 8, 0.5) # start, stop-before, step
n.dtype # float64
n # numpy arrays and arange must be all the same data type
```

```python
# we may use lists or tuples to make arrays
l1 = [3,4,5]
l2 = [9,8,7]
m = np.array([l1, l2])
m # there is no limit to the nubmer of dimensions of an array
```

```python
# as an example here is a 6-d array
e = np.random.randn(2,3,4,5,2,3)
e
```

```python
# creating empty arrays
f = np.empty( (3,6) ) # make an array in some un-used memory
f # CAREFUL - anything left in memory will be arbitrarily used
f = np.zeros( (3,6), 'float64' ) # make an array in some un-used memory AND set to zero
f
```

```python
# We may alter the dtype of the entire collection
s = f.astype(np.str_) # operates quickly across the entire array
s
```

```python
# Operations across data structures
t = np.random.randn(9000,9000)
corrected = (t+3)**3 # we may choose to apply a function here
corrected # Functional operations are extremely performant
```

##### Slicing

```python
corrected[0:5] # [start:stop-before]
corrected[::-1] # the entire reverse array
```

```python
# Numpy structures are mutable
corrected[0,2] = 99 # NB the dtype will be the same as the rest of the array
corrected[2,0] = 0
corrected.shape
corrected.T
```
